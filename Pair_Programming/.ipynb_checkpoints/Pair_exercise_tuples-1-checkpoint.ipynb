Creating a tuple looks weirdly unstructured

Note that the items in a tuple can be of any type

```python
a=1,2,"bob",3,"cathy"
```

```python
a
```

```python
# comprehension

[print(val) for val in a]
```

```python
#indexing

a[1]
```

```python
#tuples are immutable, no changes allowed

a[1]=3
```

```python
# we can combine tuples, since we can put anything we want in a typle

b=a,3,4,5

b
```

```python
#b is a tuple
# the first item in b is a tuple
b[0]
```

```python
len(b)
```

```python
#tuple unpacking

w,x,y,z=b

print(w)
print(x)
print(y)
print(z)
```
